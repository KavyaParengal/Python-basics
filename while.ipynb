```python
n=int(input('Enter the value of n : '))
sum=0
i=1
while(i<=n):
  sum=sum+i
  i=i+1
print('sum of ',n, 'numbers = ',sum)
```

```python
num=int(input('Enter a number : '))
fact=1
i=1
while(i<=num):
  fact=fact*i
  i=i+1
print('Factorial of ',num,' = ',fact)
```

```python
num=int(input('Enter the limit : '))
print('Even numbers upto the limit ',num)
i=1
while(i<=num):
  if(i%2==0):
    print(i)
  i=i+1
```
