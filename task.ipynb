1.program to print multiplication table of a given numbers

```python
num=int(input('Enter a number : '))
for i in range(1,11):
  mult=i*num
  print(i,' * ',num,' = ',mult)
```

3.print sum of dictionary values

```python
d1={
    'a':2,
    'b':4,
    'c':16,
    'd':25,
    'e':30
}
val=d1.values()
print(val)
sum=0
for i in val:
  sum=sum+i
print(sum)
```

5.program to print if a given year is leap year or not




```python
yr=int(input('Enter a year : '))
if((yr%4==0) or (yr%400==0) and (yr%100!=0)):
  print(yr,' is a leap year . ')
else:
  print(yr,' not a leap year')
```
